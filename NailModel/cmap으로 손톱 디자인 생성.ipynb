```python
import cv2, warnings , os
import numpy as np
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
```

```python
mk = np.random.uniform(0, 200, (128,128))
plt.imshow(mk, cmap='gray');
```

```python
from scipy.ndimage.filters import gaussian_filter
mode = ['reflect', 'constant','nearest', 'wrap']
plt.figure(figsize = (12, 3))
for i, val in enumerate(mode):
    plt.subplot(1, 4, i+1)
    mk_gs = gaussian_filter(mk, 10, mode= val)
    plt.title(val, fontsize=15); plt.axis('off')
    plt.imshow(mk_gs, cmap='PiYG');
```

```python
plt.figure(figsize = (15, 3))
for i in range(5):
    plt.subplot(1, 5, i+1)
    mk = np.random.uniform(0, 200, (128,128))
    mk_gs = gaussian_filter(mk, 40, mode= 'nearest')
    plt.title(i+1, fontsize=15); plt.axis('off')
    plt.imshow(mk_gs, cmap='PiYG');
```
