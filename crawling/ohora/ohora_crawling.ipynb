```python
from urllib.request import urlopen
from bs4 import BeautifulSoup as bs
import time

img_list = []
name_list = []

page_nums = [ i for i in range(1,23) ]
for i, page_num in enumerate(page_nums):
    url = "https://www.ohora.kr/product/list.html?cate_no=160&page_num="+str(page_num)
    time.sleep(5)
    html = urlopen(url)
    soup = bs(html, "html.parser")
    list_soup = soup.find('ul', id="prd_ori_id")
    # print(list_soup)
    base_soup = list_soup.find_all('li')
    # print(base_soup)
    # print(f"{page_num}페이지 시작")
    # print(url)
    # print(len(base_soup))
    
    for base in base_soup:
        # 이미지
        imgs= base.find_all('img')
        img_src = []
        for img in imgs:
            try:
                img_src.append(img.attrs['data-src'])
            except:
                pass
        img_list.append(img_src)
        
        # 제목
        mask_soup = base.find('div', class_="base_mask")
        name_soup = mask_soup.find_all('p', class_="name")
        name_list.append(name_soup[0].text)
        print(name_soup[0].text)
        time.sleep(0.5)

    print(f"{page_num}페이지 끝\n")
```

```python
len(img_list),len(name_list)
```

```python
import pandas as pd
ohora_df = pd.DataFrame({"img": img_list,
                         "name": name_list})
```
